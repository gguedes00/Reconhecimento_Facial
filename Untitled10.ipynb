```python
!pip install mtcnn opencv-python numpy keras-facenet scikit-learn

import os
import cv2
import numpy as np
from mtcnn import MTCNN
from keras_facenet import FaceNet
from sklearn.neighbors import NearestNeighbors
from google.colab import files
from google.colab.patches import cv2_imshow

# ========== CONFIGURA√á√ÉO DO DATASET ==========
def criar_dataset_gabriel():
    !rm -rf dataset
    !mkdir -p dataset/Gabriel  # Pasta fixa para o Gabriel

    print("""
    üé¨ PASSO 1/3: CRIAR DATASET DO GABRIEL

    Fa√ßa upload de 3-5 fotos do Gabriel:
    - Rosto frontal bem vis√≠vel
    - Boa ilumina√ß√£o
    - Formatos aceitos: JPG, PNG, JPEG
    """)

    uploaded = files.upload()
    detector = MTCNN()
    count = 0

    for filename in list(uploaded.keys()):
        try:
            img = cv2.imread(filename)
            img = cv2.resize(img, (800, 600))
            rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            # Detectar maior rosto
            detections = detector.detect_faces(rgb)
            if detections:
                main_face = max(detections, key=lambda x: x['box'][2]*x['box'][3])
                x, y, w, h = main_face['box']

                if w < 100 or h < 100:
                    print(f"‚ö†Ô∏è {filename}: Rosto muito pequeno ({w}x{h}px)")
                    continue

                face = rgb[y:y+h, x:x+w]
                face = cv2.resize(face, (160, 160))
                cv2.imwrite(f'dataset/Gabriel/face_{count}.jpg', cv2.cvtColor(face, cv2.COLOR_RGB2BGR))
                count += 1

                # Mostrar preview
                cv2.rectangle(rgb, (x,y), (x+w,y+h), (0,255,0), 2)
                cv2_imshow(cv2.resize(rgb, (300,300)))
                print(f"‚úÖ {filename}: Aceita")
            else:
                print(f"‚ùå {filename}: Nenhum rosto detectado")

        except Exception as e:
            print(f"‚ö†Ô∏è Erro em {filename}: {str(e)}")

    if count < 3:
        raise ValueError(f"‚ö†Ô∏è Insuficientes {count} fotos v√°lidas. M√≠nimo 3 necess√°rias")

    print(f"\n‚úî Dataset do Gabriel criado com sucesso! {count} fotos")

# ========== SISTEMA DE RECONHECIMENTO ==========
class ReconhecedorGabriel:
    def __init__(self):
        self.detector = MTCNN()
        self.embedder = FaceNet()
        self.knn = NearestNeighbors(n_neighbors=1, metric='cosine')
        self.labels = []
        self.embeddings = []

    def treinar_modelo(self):
        print("\nüîß Treinando modelo para reconhecer o Gabriel...")
        self.labels = []
        self.embeddings = []

        # Processar apenas a pasta do Gabriel
        for foto in os.listdir('dataset/Gabriel'):
            img = cv2.imread(os.path.join('dataset/Gabriel', foto))
            if img is not None:
                try:
                    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                    embedding = self.embedder.embeddings([rgb])[0]
                    self.embeddings.append(embedding)
                    self.labels.append("Gabriel")  # Nome fixo
                except:
                    continue

        if len(self.embeddings) < 3:
            raise ValueError("‚ö†Ô∏è Dados insuficientes para treinamento")

        X = np.array(self.embeddings)
        self.knn.fit(X)
        print(f"‚úÖ Treinamento conclu√≠do com {X.shape[0]} amostras")

    def reconhecer(self, caminho_imagem):
        img = cv2.imread(caminho_imagem)
        if img is None:
            raise FileNotFoundError("Imagem n√£o encontrada")

        img = cv2.resize(img, (800, 600))
        rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        detections = self.detector.detect_faces(rgb)

        for det in detections:
            try:
                x, y, w, h = det['box']
                x, y = abs(x), abs(y)
                face = rgb[y:y+h, x:x+w]
                face = cv2.resize(face, (160, 160))

                embedding = self.embedder.embeddings([face])[0]
                distancia, indice = self.knn.kneighbors([embedding])

                rotulo = "Gabriel" if distancia[0][0] < 0.5 else "Desconhecido"
                cor = (0,255,0) if rotulo == "Gabriel" else (0,0,255)

                cv2.rectangle(img, (x,y), (x+w,y+h), cor, 2)
                cv2.putText(img, f"{rotulo} ({distancia[0][0]:.2f})",
                            (x,y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, cor, 2)
            except:
                continue

        cv2_imshow(img)
        return img

# ========== EXECU√á√ÉO ==========
if __name__ == "__main__":
    criar_dataset_gabriel()

    rg = ReconhecedorGabriel()
    rg.treinar_modelo()

    print("\nüéØ PASSO 3/3: TESTE O SISTEMA")
    print("Fa√ßa upload de uma foto para reconhecimento:")
    files.upload()

    try:
        extensoes = ('.jpg', '.jpeg', '.png', '.webp')
        arquivos = [f for f in os.listdir() if f.lower().endswith(extensoes)]

        if not arquivos:
            raise ValueError("Nenhuma imagem v√°lida encontrada")

        test_img = arquivos[0]
        print(f"\nüîç Analisando: {test_img}")
        rg.reconhecer(test_img)

    except Exception as e:
        print(f"""
        ‚ö† ERRO: {str(e)}
        Solu√ß√µes:
        1. Verifique o formato do arquivo
        2. Confira se o upload foi conclu√≠do
        3. Tente novamente com outra imagem
        """)
```
